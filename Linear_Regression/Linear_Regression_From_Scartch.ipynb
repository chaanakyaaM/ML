# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Import data:
| x                    | y |
| ---------------------| ---------|
| 1                    | 2        |
| 2                    | 4        |
| 3                    | 6        |
| 4                    | 8        |
| 5                    | 10        |


```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

## Assign random values to Q0, Q1, alpha(learning rate)

```python
Q0 = 0
Q1 = 1
alpha = 0.1
```

## Define Hypothesis

```python
def predict(Q0, Q1, X):
    return Q0 + Q1 * X
```

## Define Error Function

```python
def error_function(predicted_value, actual_value):
    return predicted_value - actual_value
```

## Define Gradient Descent Function

```python
def gradient_descent(alpha, Q0, Q1, x, y, epoches):
    m = len(x)
    for i in range(epoches):
        # Store old values for logging (optional)
        v = Q0
        w = Q1

        # Step 1: Predict using current Q0 and Q1
        y_pred = predict(Q0, Q1, x)

        # Step 2: Calculate error
        error = error_function(y_pred, y)

        # Step 3: Calculate gradients
        gradient_Q0 = (1/m) * np.sum(error)        # derivative w.r.t Q0
        gradient_Q1 = (1/m) * np.sum(error * x)    # derivative w.r.t Q1

        # Step 4: Update Q0 and Q1 using the gradients
        Q0 = Q0 - alpha * gradient_Q0
        Q1 = Q1 - alpha * gradient_Q1

        # Optional: Print progress
        print(f'epoch:{i} --- Q0:{Q0:.4f} --- Q1:{Q1:.4f} --- ΔQ0:{Q0 - v:.4f} --- ΔQ1:{Q1 - w:.4f}')
    
    return Q0, Q1
```

## Compute the values using the functions

```python
Q0 , Q1 = gradient_descent(alpha, Q0, Q1, x, y, 300)
```

## Visualize the line and data points

### - Visualize the data points

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y,label='Data Points')
plt.show()
```

### - Visualize the line over the data points

```python
plt.scatter(x,y)
x_line = np.linspace(0,5,100)
y_line = Q0 + Q1 * x_line
plt.plot(x_line,y_line, linestyle = '--' ,color = 'red',label=f'Regression Line: y = {Q0:0.1} + {Q1:0.2}x')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Plot of Data with Learned Linear Regression Line')
plt.legend()
plt.show()
```
