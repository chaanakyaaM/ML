# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Import data:
| x                    | y |
| ---------------------| ---------|
| 1                    | 2        |
| 2                    | 4        |
| 3                    | 6        |
| 4                    | 8        |
| 5                    | 10        |


```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

## Assign random values to Q0, Q1, alpha(learning rate)

```python
Q0 = 0
Q1 = 1
alpha = 0.1
```

## Define Hypothesis

```python
def predict(Q0, Q1, X):
    return Q0 + Q1 * X
```

## Define Error Function

```python
def error_function(predicted_value, actual_value):
    return predicted_value - actual_value
```

## Define Gradient Descent Function

```python
def gradient_descent(alpha, Q0, Q1, x, y, epoches):
    m = len(x)
    for i in range(epoches):
        v = Q0
        w = Q1
        Q0 = Q0 - alpha * (1/m) * np.sum(error_function(predict(Q0, Q1, x), y) * 1)  # Multiplying with 1 because X0 = 1
        Q1 = Q1 - alpha * (1/m) * np.sum(error_function(predict(Q0, Q1, x), y) * x)  # X1 = x
        print(f'epoch:{i} --- Q0:{Q0} --- Q1:{Q1} --- rate of change = {Q0 - v}')
    return Q0, Q1
```

## Compute the values using the functions

```python
Q0 , Q1 = gradient_descent(alpha, Q0, Q1, x, y, 300)
```

## Visualize the line and data points

### - Visualize the data points

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y,label='Data Points')
plt.show()
```

### - Visualize the line over the data points

```python
plt.scatter(x,y)
x_line = np.linspace(0,5,100)
y_line = Q0 + Q1 * x_line
plt.plot(x_line,y_line, linestyle = '--' ,color = 'red',label=f'Regression Line: y = {Q0} + {Q1}x')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Plot of Data with Learned Linear Regression Line')
plt.legend()
plt.show()
```
