```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Boston.csv')
df.head()
```

```python
y = df['medv']
y
```

```python
x = df[df.columns[:-1]]
x
```

```python
model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=42)

model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, pred)
print(f"Mean Squared Error: {mse}")

# MSE is the average of the squared differences between the predicted and actual values.
# A lower MSE is better. It indicates that, on average, your model's predictions are closer to the actual values.

# Calculate R-squared (coefficient of determination)
r2 = model.score(X_test, y_test)
print(f"R-squared: {r2}")

# The R-squared value measures the proportion of the variance in the dependent variable that is predictable from the independent variables.
# It ranges from 0 to 1.
# In simple terms: accuracy.

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, pred, color='blue', alpha=0.6)
plt.title('Actual vs Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)  # Ideal line y = x
plt.show()
```
